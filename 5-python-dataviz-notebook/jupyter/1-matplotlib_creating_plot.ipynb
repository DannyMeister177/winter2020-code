We need to import the necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Now we need to enable jupyter to plot directly in the notebooks

```python
%matplotlib inline
```

Creating an array with 1 to 20 elements

```python
linear = np.arange(1, 20)
plt.plot(linear)
```

Creating an array with the squared values from previous array

```python
square = linear ** 2
plt.plot(square)
```

Creating an array with the log values from previous array

```python
log = np.log(linear)
plt.plot(log)
```

Creating an array with 20 random values

```python
random = np.random.randint(0, 100, 20)
plt.plot(random)
```

Plotting two values at the same time (using x and y axis)

```python
plt.plot(log, square)
```

Setting axes Title and labels

```python
plt.title('Linear by Square Plot')
plt.plot(linear, square)
plt.xlabel('Linear')
plt.ylabel('Square')
```
